# Making model, training and testing 

```python
import pandas as pd
df=pd.read_csv('Dataset.csv')
df=df.drop(['Unnamed: 0','username','display_url','comments_disabled','edge_media_to_caption','tags','shareable','dimensions_height','dimensions_width'],axis=1)
df
```

# Since the dataset is clubbed as per usernames it needs to be uniformly shuffled

```python
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df=df.set_index('shortcode')
```

# Generating training and testing sets

```python
from sklearn.model_selection import train_test_split
train=df.drop('edge_media_preview_like',axis=1)
test=df['edge_media_preview_like']
test
```

# Splitting dataset for training and testing purpose

```python
x_train,x_test,y_train,y_test=train_test_split(train,test,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
ans=logreg.predict(x_test)
```
