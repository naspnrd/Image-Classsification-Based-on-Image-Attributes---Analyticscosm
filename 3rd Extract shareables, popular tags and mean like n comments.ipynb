# dataset file contains comment_disabled feature of boolean type, and most likely as trend follows that controversial post have comment disabled, and publish of such images may cause temporary(shadow) or permanent ban

```python
import pandas as pd
import numpy as np
df=pd.read_csv('new_combined_csv.csv',encoding='utf8')
#df['tags']=df['tags'].apply(lambda x: x.length()>0)
#df[df['tags'].map(len)>0]
#df['tgas']=df['tgas'].tolist()
#df['tagslist'] = [','.join(map(str, l)) for l in df['tags']]
df = df[pd.notnull(df['tags'])]
df = df[pd.notnull(df['comments_disabled'])]
df=df.loc[df['comments_disabled']=='FALSE']
df=df[df['tags'].apply(lambda x: len(x) > 2)]
df['tags']=df['tags'].apply(lambda x: x[1:-1])
print(len(df['tags']))
df
```

# tags column values are compared to a set of few handpicked stopwords presence of which can cause a permanent or a shadow ban on instagram
# even if a single tag in a post is present in stopwords then it becomes unshareable 

```python
dkeys=df['shortcode'].tolist()
dtags=df['tags'].tolist()
dtags=[x.split(', ') for x in dtags]
empty=[]
for i in dtags:
    x=[q[1:-1] for q in i]
    empty.append(x)
dtags=empty
#print(empty)
#if empty[0][0]=='roboticseminar':
#    print('done')

#print(taga[0][0][1:-1])
#taga[0][0]
#d={}
#q=len(kiys)
mydict = dict(zip(dkeys, dtags))
#print(mydict)


#mydict['BemQM09A0qL']=1
#print(mydict)
stopwords = [line.rstrip('\n') for line in open('stopwords.txt')]

stopwords=stopwords[0].split(' ')
#print(stopwords)
shareable={}
for i in mydict:
    shareable[i]=1
    temp=list(mydict[i])
    #print(temp)
    #break
    #temp=['robot']
    flag=False
    for x in temp:
        #print(temp)
        for st in stopwords:
            #print(st,x)
            if(x==st):
                shareable[i]=0
                flag=True
                break;
        if flag==True:
            break;
sharing=[shareable[i] for i in shareable]

#df["comments_disabled"]=df["comments_disabled"].replace({False:0,True:1})

df['shareable']=sharing

#df['shareable']=df['shareable']*df['comments_disabled']
df=df.loc[df['shareable']==1]

df
```

# with over a millions of hashtags used everyday, a new feature in dataset can be created by counting those tags of a post that are in top 100 most popularily used hashtags

```python
dkeys=df['shortcode'].tolist()
dtags=df['tags'].tolist()
dtags=[x.split(', ') for x in dtags]
empty=[]
for i in dtags:
    x=[q[1:-1] for q in i]
    empty.append(x)
dtags=empty
#print(empty)
#if empty[0][0]=='roboticseminar':
#    print('done')

#print(taga[0][0][1:-1])
#taga[0][0]
#d={}
#q=len(kiys)
mydict = dict(zip(dkeys, dtags))
#print(mydict)


popular_tags = [line.rstrip('\n') for line in open('Popular hashtags.txt')]

popular_tags=popular_tags[0].split(' ')
#to remove duplicates
popular_tags=list(dict.fromkeys(popular_tags))
#print(popular_tags)
popularity={}
for i in mydict:
    popularity[i]=0
    temp=list(mydict[i])
    for x in temp:
        #print(temp)
        for st in popular_tags:
            if(x==st):
                popularity[i]+=1
                break;
popular_hashtags=[popularity[i] for i in popularity]

df['popular_hashtags']=popular_hashtags
df=df.set_index('username')
df.to_csv(r'shareables csv file.csv')
df
```

# The shareables csv file obtained above is manipulated to get average number of likes and comments per user depending on the data in shareables.csv
